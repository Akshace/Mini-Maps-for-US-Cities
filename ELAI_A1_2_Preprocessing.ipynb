### DATASET README

This is a simple dataset of North American (though mostly U.S.) major roads.

city-gps.txt contains one line per city, with three fields per line, 
delimited by spaces. The first field is the city, followed by the latitude,
followed by the longitude.

road-segments.txt has one line per road segment connecting two cities.
The space delimited fields are:

- first city
- second city
- length (in miles)
- speed limit (in miles per hour)
- name of highway


Note that there are mistakes and bugs in these files and your code should
still operate correctly; e.g. not all cities that appear in road-segments.txt
have a corresponding line in city-gps.txt. You should assume that all roads
in road-segments.txt are bidirectional, i.e. none are one-way roads, so
that it's possible to travel from the first city to the second city at the
same distance at speed as from the second city to the first city.

```python
class Cities:
    def __init__(self, citygps, road_segments):
        self.citygps = citygps
        self.road_segments = road_segments
        
        
    def distance_cost_heuristic(self,succ_city, goal_city):
        #calculate dist between curr and succ
        try:
            return np.sqrt(np.square(citygps[citygps['City'] == goal_city]['la'].iloc[0]-\
                                     citygps[citygps['City'] == succ_city]['la'].iloc[0])\
                        + np.square(citygps[citygps['City'] == goal_city]['lo'].iloc[0]-\
                                    citygps[citygps['City'] == succ_city]['lo'].iloc[0]))
        
        except IndexError:
            print('City not found in distance heuristic ')
            return 0
        
        
    def time_heuristic(self, curr_city, succ_city, goal_city):
        # calculate speed of  CURR and 
        try:
            return (np.sqrt(np.square(citygps[citygps['City'] == goal_city]['la'].iloc[0]-\
                                     citygps[citygps['City'] == succ_city]['la'].iloc[0])\
                        + np.square(citygps[citygps['City'] == goal_city]['lo'].iloc[0]-\
                                    citygps[citygps['City'] == succ_city]['lo'].iloc[0])))/\
                    (road_segments[(road_segments['first_city'] == curr_city)\
              & (road_segments['second_city'] == succ_city)]['speed_limit'].iloc[0])
        
        except IndexError:
            print('City not found in time heuristic ')
            return 0
        
    def time_cost(self, curr_city, succ_city):
        # calculate speed of  CURR and 
        try:
            return (np.sqrt(np.square(citygps[citygps['City'] == curr_city]['la'].iloc[0]-\
                                     citygps[citygps['City'] == succ_city]['la'].iloc[0])\
                        + np.square(citygps[citygps['City'] == curr_city]['lo'].iloc[0]-\
                                    citygps[citygps['City'] == succ_city]['lo'].iloc[0])))/\
                    (road_segments[(road_segments['first_city'] == curr_city)\
              & (road_segments['second_city'] == succ_city)]['speed_limit'].iloc[0])
        
        except IndexError:
            print('City not found in time heuristic ')
            return 0
```

```python
c = Cities(citygps, road_segments)
c.distance_cost_heuristic('Y_City,_Arkansas', 'Abbot_Village,_Maine')
```
